## Create a query to answer the following questions:

### How many tables are in the austin_bikeshare dataset?

```python
query = """
SELECT * FROM bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.TABLES;
"""
query_job = client.query(query).to_dataframe()
query_job
```

### How many columns in the bikeshare_stations table have STRING data?

```python
query = """
SELECT * FROM bigquery-public-data.austin_bikeshare.INFORMATION_SCHEMA.COLUMNS
WHERE
  TABLE_NAME = "bikeshare_stations" and
  DATA_TYPE = "STRING"
"""
query_job = client.query(query).to_dataframe()
query_job
```
