```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time

driver = webdriver.Chrome()
df = pd.DataFrame()

coin_name = []
coin_price = []
coin_volume = []
coin_supply = []
for i in range (10):
    url = f"https://coinmarketcap.com/?page={i+1}"
    driver.get(url)
    scroll_pause_time= 1
    screen_height = driver.execute_script("return window.screen.height;")
    i = 1
    while True:
        driver.execute_script(f"window.scrollTo(0, {screen_height * i});")
        i += 1
        time.sleep(scroll_pause_time)
        scroll_height = driver.execute_script("return document.body.scrollHeight;")
        if screen_height * i > scroll_height:
            break


    html = driver.page_source

    soup = BeautifulSoup(html, "html.parser")
    #print(soup.prettify())

    result = soup.find('tbody')

    #print (result)
    #Method 1
    

    for data in result:
        #print (result)
        c_name = data.find("p", {'class':'sc-71024e3e-0 ehyBa-d'})
        if c_name != None:        
            coin_name.append (c_name.get_text())
        else:
            coin_name.append(None)
        
        c_price = data.find("div", {'class':'sc-b3fc6b7-0 dzgUIj'})
        if c_price != None:
            coin_price.append (c_price.get_text())
        else:
            coin_price.append (None)

        c_volume = data.find("p", {'class':'sc-71024e3e-0 bbHOdE font_weight_500'})
        if c_volume != None:
            coin_volume.append (c_volume.get_text())
        else:
            coin_volume.append(None)

        c_supply = data.find("p", {'class':'sc-71024e3e-0 hhmVNu'})
        if c_supply != None:
            coin_supply.append (c_supply.get_text())
        else:
            coin_supply.append(None)
    
    

    #df['coin_name'] = [coin_name.get_text() for coin_name in i.find("p", {'class':'sc-71024e3e-0 ehyBa-d'})]

df['coin_name'] = coin_name
df['coin_price'] = coin_price
df['coin_volume'] = coin_volume
df['coin_supply'] = coin_supply

df

```
