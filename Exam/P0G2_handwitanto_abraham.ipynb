### 2. DATA CLEANING

#### Memperbaiki nama kolom
- membuat kolom name menjadi lowercase
- menghapus whitespace dan menambahkan "_" pada kolom yang membutuhkan

````python
#syntax ini digunakan untuk mengganti nama kolom :
# - menjadi lowercase
# - menghilangkan whitespace
# - menambahkan '_' untuk nama kolom dengan 2 kata atau lebih
df = df.rename(columns={'Segment' : 'segment', 'Country' : 'country', ' Product ' : 'product', 
                   ' Discount Band ' : 'discount_band', ' Units Sold ' : 'units_sold',
                    ' Manufacturing Price ' : 'manufacturing_price', ' Sale Price ' : 'sale_price', 
                    ' Gross Sales ' : 'gross_sales', ' Discounts ' : 'discounts',
                    '  Sales ': 'sales', ' COGS ' : 'cogs', ' Profit ' : 'profit', 'Date' : 'date', 
                    'Month Number' : 'month_number', ' Month Name ' : 'month_name', 'Year' : 'year'})
````

#### Menghapus special character pada values

````python
#function ini gitunakan untuk mengganti value dari seluruh dataframe yang memiliki simbol
# '$' dengan sebelumnya menggunakan function 'isinstance()' untuk memastikan value yang
# diubah memiliki tipe data string, jika benar, maka hapus simbol '$'
def hapus_dollar(value):
    if isinstance(value, str):
        return value.replace('$', '')
    return value
#function ini gitunakan untuk mengganti value dari seluruh dataframe yang memiliki simbol
# ',' dengan sebelumnya menggunakan function 'isinstance()' untuk memastikan value yang
# diubah memiliki tipe data string, jika benar, maka hapus simbol ','
def hapus_koma(value):
    if isinstance(value, str):
        return value.replace(',', '')
    return value

#syntax ini digunakan untuk menghapus simbol '-' pada kolom 'discounts' dan mengganti jadi string '0'
df.discounts = df.discounts.str.replace('-', '0')
#syntax ini digunakan untuk menghapus simbol '(' dan ')' pada kolom profit
df.profit = df.profit.str.replace('(', '')
df.profit = df.profit.str.replace(')', '')
#syntax ini digunakan untuk menghapus simbol '-' pada kolom 'profit' dan mengganti jadi string '0'
df.profit = df.profit.str.replace('-', '0')
#syntax ini digunakan untuk menghapus whitespace pada kolom 'product' dan 'discount_band'
df['product'] = df['product'].str.replace(' ', '')
df.discount_band = df.discount_band.str.replace(' ', '')
#syntax ini digunakan untuk menjalankan function 'hapus_dollar' dan 'hapus_koma'
df = df.map(hapus_dollar)
df = df.map(hapus_koma)

df
````

#### Mengubah tipe data menjadi numerik

````python
#syntax ini digunakan untuk mengubah tipe data pada kolom tersebut menjadi numeric (float64),
#penggunaan tipe data float karena value di dalamnya memiliki bilangan desimal
df[['units_sold','manufacturing_price','sale_price','discounts',
    'gross_sales','sales','cogs','profit']] = df[['units_sold','manufacturing_price',
    'sale_price','discounts','gross_sales','sales','cogs',
    'profit']].apply(pd.to_numeric)

#syntax ini untuk mengubah tipe data pada kolom 'date' menjadi 'datetime'
df.date = df.date.apply(pd.to_datetime)
df.info()
````

### 3. DATA NORMALIZATION
#### Membuat Dataframe baru untuk Segment, Product, Country, dan Discount Band

````python
#syntax ini digunakan untuk membuat dataframe baru, dengan tujuan normalization, 
#sehingga dari dataframe awal, di partisi menjadi 5 dataframe.
#syntax ini juga melakukan copy value dari kolom terkait di dataframe awal
df_segment = pd.DataFrame(df['segment'])
df_country = pd.DataFrame(df['country'])
df_product = pd.DataFrame(df[['product','manufacturing_price']])
df_discount_band = pd.DataFrame(df['discount_band'])

#syntax ini berfungsi untuk menghapus data duplikat pada ke-4 dataframe yang baru dibuat,
#agar hanya tersisa value yang unik
df_segment.drop_duplicates(inplace=True)
df_product.drop_duplicates(inplace=True)
df_country.drop_duplicates(inplace=True)
df_discount_band.drop_duplicates(inplace=True)

#syntax ini digunakan untuk membuat kolom baru berupa ID, yang nantinya 
#akan digunakan sebagai FOREIGN KEY, dan terhubung dengan dataframe awal
df_segment['segment_id'] = ['SG01','SG02','SG03','SG04','SG05']
df_product['product_id'] = ['P01','P02','P03','P04','P05','P06']
df_country['country_id'] = ['C01','C02','C03','C04','C05']
df_discount_band['discount_band_id'] = ['DB01','DB02','DB03','DB04']

#syntax ini berfungsi untuk mengganti value pada kolom segmen di dataframe awal
#disesuaikan dengan value segmen_id pada dataframe df_segment
segment_v = {
    'Government': 'SG01',
    'Midmarket':'SG02',
    'Channel Partners':'SG03',
    'Enterprise':'SG04',
    'Small Business':'SG05'
}

df['segment'] = df.segment.replace(segment_v)

#syntax ini berfungsi untuk mengganti value pada kolom produk di dataframe awal
#disesuaikan dengan value produk_id pada dataframe df_produk
product_v = {
    'Carretera':'P01',
    'Montana':'P02',
    'Paseo':'P03',
    'Velo':'P04',
    'VTT':'P05',
    'Amarilla':'P06'
}
df['product'] = df['product'].replace(product_v)

#syntax ini berfungsi untuk mengganti value pada kolom country di dataframe awal
#disesuaikan dengan value country_id pada dataframe df_country
country_v = {
    'Canada':'C01',
    'Germany':'C02',
    'France':'C03',
    'Mexico':'C04',
    'United States of America':'C05'
}
df['country'] = df['country'].replace(country_v)

#syntax ini berfungsi untuk mengganti value pada kolom discount_band di dataframe awal
#disesuaikan dengan value discount_band_id pada dataframe df_discount_band
discount_band_v = {
    'None':'DB01',
    'Low':'DB02',
    'Medium':'DB03',
    'High':'DB04'
}
df['discount_band'] = df['discount_band'].replace(discount_band_v)

#syntax ini berfungsi menambahkan kolom "finance_id" pada dataframe awal
#menambahkan value dimulai dari 1 hingga akhir jumlah row, 
# dan membuat kolom finance_id sebagai PRIMARY KEY (Index)
df['finance_id'] = range(1, 1+len(df))
df.set_index('finance_id', inplace=True)

#syntax ini berfungsi membuat kolom ID pada masing-masing dataframe yang baru dibuat
#menjadi index agar dapat terkoneksi dengan dataframe awal
df_segment.set_index('segment_id', inplace=True)
df_country.set_index('country_id', inplace=True)
df_product.set_index('product_id', inplace=True)
df_discount_band.set_index('discount_band_id', inplace=True)

#syntax ini berfungsi mengubah nama kolom pada dataframe awal untuk menjadi 
#FOREIGN KEY 
df = df.rename(columns={'segment' : 'segment_id','country':'country_id',
                        'product':'product_id','discount_band':'discount_band_id' })

df
````

#### Menghapus Column yang tidak diperlukan

````python
#syntax ini digunakan untuk menghapus kolom yang tidak digunakan dengan alasan : 
# - Kolom month_number, month_name, dan year memiliki value yang sama dengan kolom date
# - kolom manufacturing_price terikat dengan df_produk, sehingga sudah terimpan pada dataframe terpisah
df.drop(columns=['month_number','month_name','year','manufacturing_price'], inplace=True)
df
````

#### Export ke CSV

````python
#syntax ini digunakan untuk melakukan convert dataframe yang sudah dibuat beserta datanya
#menjadi file .csv untuk selanjutnya dibuat kedalam database 
df.to_csv('finance.csv', index=False)
df_segment.to_csv('segment.csv', index=True)
df_country.to_csv('country.csv', index=True)
df_product.to_csv('product.csv', index=True)
df_discount_band.to_csv('discount_band.csv', index=True)
````
