```python
# load data
df = pd.read_csv('https://raw.githubusercontent.com/FTDS-learning-materials/phase-0/main/src/teams.csv')
```

```python
df['new_col'] = 1
```

```python
df['new2'] = df.id * 5
```

```python
df
```

```python
df._append({'division': 'One', 'conference': 'Meja Bundar', 'school_name':'ABC'}, ignore_index=True)
```

```python
df.drop(columns='new_col', inplace=True)
```

```python
df = df[~(df.division == 'One')]
```

```python
df
```

```python
df.groupby('conference')['id'].agg(['mean', 'sum', 'count'])
```

```python
df[['id', 'new2']].agg({'id': 'sum', 'new2': 'mean'})
```
