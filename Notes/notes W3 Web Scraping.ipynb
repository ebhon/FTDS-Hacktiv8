```python
from bs4 import BeautifulSoup
from selenium import webdriver

title = []
price = []

# membuka browser
driver = webdriver.Chrome()

# scraping banyak halaman
# 0,1,2
for i in range(1):
    url = "https://www.gramedia.com/categories/buku?page=" + str(i+1)

    # membuka website
    driver.get(url)

    # mengakses atribute
    html = driver.page_source

    # membuat object BS4
    page = BeautifulSoup(html, "html.parser")

    # tampilkan data HTML yang discraping
    # print(page.prettify())

    # method untuk scraping (mencari element)
    # .find()               -> menemukan 1 element -> walaupun ada banyak element
    # .find_all()           -> menemukan > 1 element -> list of element

    # parameter find/find_all
    # 1. nama element yang dicari -> string
    # 2. (optional) atribute dari element yang dicari -> dictionary

    # method .get_text() -> untuk mengekstraksi data yang ada di dalam element (antara tag buka dan tag tutup)

    # teknik scraping
    # 1. scraping setiap element menggunakan find/find_all
    # 2. scraping 1x untuk elemen yang kita inginkan

    
    # <div _ngcontent-web-gramedia-c26="" class="list-title">Muhammad The Messenger: Periode Mekah</div> -> get_text()
                                                                        data
    # <img src="http://www.google.com/index.jpg"> -> sama seperti memanggil key dalam dictionary
                            data
    
    
    
    
    
    result = page.find_all("div", {"class":"ng-star-inserted"})

    for entity in result:
        # extrak judul
        elementTitle = entity.find('div', {"class":"list-title"})
        # jika elementTitle -> None -> gabisa panggil .get_text() -> Error
        if elementTitle != None:
            title.append(elementTitle.get_text())
        else: 
            title.append(None)

        # extrak harga
        # price.append(entity.find('p', {"class":"formats-price"}).get_text())

        elementImg = entity.find("img", {"class":"product-list-img"})
        if elementImg != None:
            print(elementImg['src'])
```

```python
import pandas as pd

df = pd.DataFrame()

df['Title'] = title
df['Price'] = price
```
