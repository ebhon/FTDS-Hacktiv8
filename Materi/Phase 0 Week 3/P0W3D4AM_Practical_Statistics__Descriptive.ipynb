## Kinds of Data

1. **Data Numerik/Kuantitatif**:
   - **Kontinu**: Data kontinu bisa berupa apa saja yang dapat diukur dan tidak terbatas pada angka bulat, seperti berat, tinggi, atau waktu. Seperti mengukur air dengan gelas ukur, mengukur tinggi badan, harga, jarak, berat, dan masih banyak lagi. Bisa `bilangan bulat` (**integer**) atau `desimal` (**float**). 
   
   - **Diskrit**: Data diskrit hanya berupa `bilangan bulat` (**integer**) dan tidak dapat dibagi, seperti jumlah buku, orang, atau mobil.

   - **Ordinal**: Data ordinal menggambarkan kategori yang `memiliki urutan atau tingkatan`, tetapi jarak antar tingkat tidak tentu. Contohnya adalah survei kepuasan dengan opsi "sangat puas," "puas," "netral," "tidak puas," dan "sangat tidak puas"
   
   - **Nominal**: Data nominal menggambarkan kategori **`tanpa`** `memiliki urutan atau tingkatan`. Contohnya adalah jenis kelamin, warna, atau nama tempat. Ini seperti label yang ditempelkan pada benda untuk menggambarkan apa itu, tanpa memberi peringkat atau urutan.

# Measurement of Dispersion

To determine the spreadness or variability of data, we can utilize various statistical measures. Here are a few common methods

*Untuk menentukan persebaran atau variabilitas data, kita dapat menggunakan berbagai ukuran statistik. Berikut adalah beberapa metode yang umum digunakan*

## Range

The range provides a simple way to assess the spread of data. It is calculated by subtracting the smallest value from the largest value in the dataset. However, the range only considers the extreme values and may not capture the full picture of variability

*Rentang menyediakan cara sederhana untuk menilai persebaran data. Dihitung dengan mengurangkan nilai terkecil dari nilai terbesar dalam kumpulan data. Namun, kisaran tersebut hanya mempertimbangkan nilai ekstrem dan mungkin tidak menangkap gambaran variabilitas secara utuh*

```python
# Langkah tersedikit
sleep['Daily Steps'].min()
```

```python
# Langkah terbanyak
sleep['Daily Steps'].max()
```

```python
# Menghitung rentang
range = sleep['Daily Steps'].max() - sleep['Daily Steps'].min()

# Cara 1: Langsung rentang
print('Range of daily steps:',range, 'steps')
# Cara 2: Memberitahu nilai min & max 
print('Range of daily steps from', sleep['Daily Steps'].min(), "to", sleep['Daily Steps'].max(), 'step')
```

## Inter Quartile Range (IQR)

The IQR measures the spread of the central portion of the data. It is the difference between the third quartile (75th percentile) and the first quartile (25th percentile) of the dataset. By focusing on the middle 50% of the data, the IQR is less sensitive to extreme values

*IQR mengukur persebaran bagian tengah data. Ini adalah perbedaan antara kuartil ketiga (Q3) dan kuartil pertama (Q1) dari kumpulan data. Dengan berfokus `median` (Q2), IQR kurang sensitif terhadap nilai ekstrem*

```python
# Mencari nilai Q1, Q2, & Q3
q1 = sleep['Daily Steps'].quantile(0.25)
q2 = sleep['Daily Steps'].quantile(0.5)
q3 = sleep['Daily Steps'].quantile(0.75)
# Menghitung nilai IQR (Q3 - Q1)
iqr = q3 - q1
# Lower bound.
low_bound = q1 - 1.5 * iqr
# Upper bound.
up_bound = q3 + 1.5 * iqr

print('Q1:',q1)
print('Q2:',q2)
print('Q3:',q3)
print('Interquartile Range (IQR):',iqr)
print('Lower Boundaries:', low_bound)    # Low Bound = Q1 - k*IQR
print('Upper Boundaries:', up_bound)    # Up Bound  = Q3 + k*IQR
print('Max. Value:', sleep['Daily Steps'].max())
print('Min. Value:', sleep['Daily Steps'].min())
```

```python
# Mencari jumlah langkah kurang dari batas bawah
sleep[sleep["Daily Steps"]<low_bound]
```

```python
# Mencari jumlah langkah lebih dari batas atas
sleep[sleep["Daily Steps"]>up_bound]
```

```python
#@title Illustration of Quartile on Boxplot
plt.figure(figsize=(6, 4))
plt.boxplot(sleep['Daily Steps'], vert=False, patch_artist=True)

plt.annotate(f'Q1 = {q1}', xy=(q1, 1.06), xytext=(q1, 1.2), arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate(f'Q2 = {q2}', xy=(q2, 1.06), xytext=(q2, 1.3), arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate(f'Q3 = {q3}', xy=(q3, 1.06), xytext=(q3, 1.2), arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate(f'IQR = {iqr}', xy=((q1+q3)/2, 0.85), ha='center')

plt.axvline(q1, linestyle='--', color='gray', label='Q1')
plt.axvline(q3, linestyle='--', color='gray', label='Q3')
plt.axvspan(q1, q3, alpha=0.2, color='gray', label='IQR')

plt.xlabel('Steps')
plt.title('Boxplot of Daily Steps')

plt.show()
```

*Simpangan baku adalah akar kuadrat dari varians. Ini mewakili jumlah tipikal atau rata-rata dimana titik data menyimpang dari rata-rata. Ini adalah ukuran persebaran yang banyak digunakan karena unitnya sama dengan data asli, sehingga lebih mudah untuk diinterpretasikan*

## Normality Test

### Testing

#### DAgostinos K^2 Test

This test combine skewness and kurtosis values to validate a data is normal or not. We will use condition that if `p-value` more than 0.05, the data has normal distribution.

- Tes ini menggunakan skewness dan kurtosis dari data untuk menguji normalitas.
- D'Agostino's K^2 Test lebih cocok untuk sampel yang lebih besar karena ia mengandalkan estimasi parameter yang lebih stabil di sampel yang lebih besar.
- Skewness mengukur simetri dalam distribusi data, sementara kurtosis mengukur seberapa berat ekor distribusi data. Dalam kata lain, skewness mengukur normal atau tidaknya persebaran data, lalu kurtosis mengukur seberapa normal data yang terdistribusi normal.
- Sama seperti Shapiro-Wilk, tes ini menghasilkan nilai p, dan jika nilai p tersebut di bawah tingkat signifikansi yang ditentukan (misalnya 0.05 atau 5%), maka akan menolak hipotesis nol (H0) bahwa data terdistribusi secara normal.

```python
res = stats.normaltest(student_perform['math score'])
print(f'p-value: {res.pvalue}')
```

#### Shapiro-Wilk Test

This testing tool is another way to test the normality of a data. The condition is the same as D'Agostino K^2 Test.

- Tes yang sangat populer untuk normalitas karena dikenal karena keakuratannya dan keandalannya pada sampel yang kecil.
- Biasanya menggunakan tes Shapiro-Wilk ketika kita memiliki set data yang relatif kecil, biasanya di bawah 50 sampel.
- Tes ini sangat sensitif terhadap penyimpangan dari distribusi normal, yang berarti ia bisa mendeteksi penyimpangan yang kecil.
- Tes ini akan memberikan Anda nilai p, dan jika nilai p tersebut lebih kecil dari tingkat signifikansi yang ditetapkan (misalnya 0.05 atau 5%), maka akan menolak hipotesis nol (H0) bahwa data terdistribusi secara normal.

```python
res = stats.shapiro(student_perform['math score'])
print(f'p-value: {res.pvalue}')
```

Unfortunately, our `math score` data has not normal distribution, but visually shows a different insight. We need to perform other tests.

## What is Outliers and How are They Important?

Outliers are data points that significantly differ from the rest of the observations in a dataset. These are values that are unusually high or low compared to the majority of the data. Outliers can arise due to various reasons, such as measurement errors, data entry mistakes, or genuine extreme observations

*Outliers adalah titik data yang berbeda secara signifikan dari observasi lainnya dalam kumpulan data. Ini adalah nilai yang sangat tinggi atau sangat rendah dibandingkan dengan sebagian besar data. Outliers dapat muncul karena berbagai alasan, seperti kesalahan pengukuran, kesalahan entri data, atau pengamatan ekstrim yang sebenarnya*

*Memahami dan mengidentifikasi outlier penting karena beberapa alasan:*
- *Outliers mungkin menunjukkan kesalahan atau anomali dalam proses pengumpulan atau pencatatan data*
- *Outliers dapat berdampak besar pada ukuran statistik seperti mean dan deviasi standar*
- *Banyak model statistik yang berasumsi bahwa data berdistribusi normal atau mengikuti distribusi tertentu*
- *Outliers dapat mempunyai pengaruh yang tidak proporsional terhadap hasil analisis statistik*
- *Dalam situasi tertentu, outlier mungkin mewakili kasus unik atau kritis yang memerlukan perhatian atau penyelidikan khusus*
- *Outliers dapat mempengaruhi interpretasi data dan proses pengambilan keputusan*

## How to Detect

### Normal Distribution: Z-Score

```python
# Mencari nilai rerata dan std
avg = student_perform['math score'].mean()
std = student_perform['math score'].std()
# Mencari nilai batas atas & batas bawah
up_bound = avg + 3*std  # mean + k*std
low_bound = avg - 3*std # mean - k*std

print(f'Upper Boundary: {up_bound:.2f}' )
print(f'Lower Boundary: {low_bound:.2f}')
```

```python
print('Nilai minimum :', student_perform['math score'].min())
print('Nilai maksimum:', student_perform['math score'].max())
```

```python
# OUTLIER -> nilainya dibawah batas bawah ATAU diatas batas atas
outlier = student_perform['math score'][(student_perform['math score']<low_bound) | (student_perform['math score']>up_bound)]

# NORMAL  -> nilainya diatas batas bawah DAN dibawah batas atas
no_outlier = student_perform['math score'][(student_perform['math score']>=low_bound) & (student_perform['math score']<=up_bound)]

print(f'Percentage of outliers: {len(outlier)*100/len(student_perform["math score"]):.2f}%')
```

### Skew Distribution: Tukey's Rule

```python
# Cek nilai Q1, Q3, dan IQR
q1 = student_perform['math score'].quantile(0.25)
q3 = student_perform['math score'].quantile(0.75)
iqr = q3-q1
# Cek nilai batas atas dan bawah
up_bound = q3 + 1.5*iqr
low_bound = q1 - 1.5*iqr

print(f'Upper Boundary: {up_bound:.2f}' )
print(f'Lower Boundary: {low_bound:.2f}')
```

```python
# OUTLIER -> nilainya dibawah batas bawah ATAU diatas batas atas
outlier = student_perform['math score'][(student_perform['math score']<low_bound) | (student_perform['math score']>up_bound)]

# NORMAL  -> nilainya diatas batas bawah DAN dibawah batas atas
no_outlier = student_perform['math score'][(student_perform['math score']>=low_bound) & (student_perform['math score']<=up_bound)]

print(f'Percentage of outliers: {len(outlier)*100/len(student_perform["math score"]):.2f}%')
```

```python
fig, ax = plt.subplots(ncols=2,figsize=(10,4))

no_outlier.plot(kind='hist',bins=20,ax=ax[0])
no_outlier.plot(kind='box',ax=ax[1])
```

The outliers has gone.

But, why it can be? it is so strange that using z-score method to measure the upper and lower boundaries, we still got outliers in our data but using tukey's rule, the outliers has gone. Meanwhile, our data distribution is normal.

This confusion can be explained that the boxplot itself constructed by quartile, which means tend to handle skewed data better than normal.

Don't worry, you are in the right path since our steps is right.

Kalau ditanya atau diminta untuk membandingkan perubahan pemusatan data (*central tendency*) sebelum dan sesudah *handle outlier* lakukan ini:

# Correlation Analysis

Correlation is a statistical measure that quantifies the relationship or association between two variables. It determines the extent to which changes in one variable are related to changes in another variable. Correlation is often expressed as a correlation coefficient, which indicates both the strength and direction of the relationship.

*Korelasi adalah ukuran statistik yang mengukur hubungan atau hubungan antara dua variabel. Ini menentukan sejauh mana perubahan dalam satu variabel berhubungan dengan perubahan dalam variabel lain. Korelasi sering kali dinyatakan sebagai koefisien korelasi, yang menunjukkan kekuatan dan arah hubungan.*



Correlation coefficients range from -1 to +1. A correlation coefficient of +1 represents a perfect positive correlation, indicating that as one variable increases, the other variable also increases proportionally. A correlation coefficient of -1 represents a perfect negative correlation, indicating that as one variable increases, the other variable decreases proportionally. A correlation coefficient of 0 indicates no linear relationship between the variables. The magnitude of the correlation coefficient indicates the strength of the relationship. Values closer to +1 or -1 indicate a stronger correlation, while values closer to 0 indicate a weaker correlation.

*Koefisien korelasi berkisar dari -1 hingga +1. Koefisien korelasi sebesar +1 menunjukkan korelasi positif sempurna yang menunjukkan bahwa jika salah satu variabel meningkat maka variabel lainnya juga meningkat secara proporsional. Koefisien korelasi sebesar -1 menunjukkan korelasi negatif sempurna, yang menunjukkan bahwa ketika satu variabel meningkat, variabel lainnya menurun secara proporsional. Koefisien korelasi sebesar 0 menunjukkan tidak adanya hubungan linier antar variabel. Besarnya koefisien korelasi menunjukkan kuatnya hubungan tersebut. Nilai yang mendekati +1 atau -1 menunjukkan korelasi yang lebih kuat, sedangkan nilai yang mendekati 0 menunjukkan korelasi yang lebih lemah.*

1. **`Pearson`**: denoted as r, measures the linear relationship between two continuous variables. It assumes that the data follows a normal distribution. It is sensitive to outliers and assumes a linear relationship, making it suitable for assessing linear associations between normally distributed continuous variables. 

*Pearson: dilambangkan dengan r, mengukur hubungan linier antara dua variabel kontinu. Diasumsikan bahwa data mengikuti distribusi normal. `Sensitif` terhadap outlier dan `mengasumsikan hubungan linier`, sehingga cocok untuk menilai hubungan linier `antara variabel kontinu` yang `terdistribusi normal`.*

2. **`Spearman`**: denoted as  (rho), measures the monotonic relationship between variables. It does not assume a specific distribution for the data and is appropriate for variables that may not have a linear relationship. It is less sensitive to outliers than the Pearson correlation and can capture non-linear relationships. It is often used when the variables have skewed distributions or when the relationship is better described by the order or ranking of the data.

*Spearman: dilambangkan dengan  (rho), mengukur hubungan monotonik antar variabel. Ini tidak mengasumsikan distribusi data yang spesifik dan sesuai untuk variabel yang `mungkin tidak memiliki hubungan linier`. Korelasi ini `kurang sensitif` terhadap outlier dibandingkan korelasi Pearson dan dapat menangkap hubungan non-linier. Hal ini sering digunakan ketika variabel memiliki `distribusi yang tidak seimbang` atau ketika hubungan lebih baik dijelaskan oleh urutan atau peringkat data.*

3. **`Kendall`**: denoted as  (tau), also measures the strength of a monotonic relationship between variables. Like the Spearman correlation, it does not assume a specific distribution for the data. It is robust to outliers and can handle non-linear relationships. It is commonly used when dealing with categorical or ordinal variables.

*Kendall: dilambangkan dengan  (tau), juga mengukur kekuatan hubungan monotonik antar variabel. Seperti korelasi Spearman, korelasi ini `tidak mengasumsikan distribusi data yang spesifik`. Ini `kuat terhadap outlier` dan dapat menangani hubungan non-linier. Ini biasanya digunakan ketika berhadapan dengan `variabel kategori atau ordinal`.*

```python
corr_r, pval_p = stats.pearsonr(sleep['Sleep Duration'], sleep['Heart Rate'])       # Pearson (r)
corr_rho, pval_s = stats.spearmanr(sleep['Sleep Duration'], sleep['Heart Rate'])    # Spearman (rho)
corr_tau, pval_k = stats.kendalltau(sleep['Sleep Duration'], sleep['Heart Rate'])   # Kendall (tau)

print(f"r-correlation  : {corr_r:.2f}, p-value: {pval_p}")      # Pearson (r)
print(f"rho-correlation: {corr_rho:.2f}, p-value: {pval_s}")    # Spearman (rho)
print(f"tau-correlation: {corr_tau:.2f}, p-value: {pval_k}")    # Kendall (tau)
```

p-value used to provide information whether the correlation is real or just by chance.

If p-value less than 0.05, so the correlation between two variables is real.

But, in our case, the correlation between sleep duration and heart rate is by chance, although based on Pearson and Spearman correlation, the show that those variables has moderate correlation.
